#Spatial Domain Filtering

Spatial Domain Filtering
A. Smoothing (Blurring)

Smoothing is a spatial domain filtering technique used to reduce noise and smooth variations in an image by averaging or modifying pixel intensities.

Types of Smoothing Filters
1. Average (Mean) Filter

Each pixel is replaced by the average of its neighboring pixels.

All pixels have equal weightage.

Effective in reducing random noise, but it blurs edges.

2. Gaussian Filter

Uses a Gaussian distribution for weighting pixels.

Central pixels have higher weightage than surrounding pixels.

Produces smoother and more natural blurring than the average filter.

Preserves edges better compared to the mean filter.

3. Median Filter

Each pixel is replaced by the median value of neighboring pixels.

Very effective for salt-and-pepper noise.

Preserves edges better since it does not average pixel values.


```python
import cv2
import matplotlib.pyplot as plt

img_avg= cv2.imread("/content/Avg_filtering.jpg")
img_gaussian= cv2.imread("/content/Guassian_filtering.png")
img_median= cv2.imread("/content/median_img.jpg")

if img_avg is None or img_gaussian is None or img_median is None:
    print("Error loading images.")
    exit()


img_avg_rgb=cv2.cvtColor(img_avg,cv2.COLOR_BGR2RGB)
img_gaussian_rgb=cv2.cvtColor(img_gaussian,cv2.COLOR_BGR2RGB)
img_median_rgb=cv2.cvtColor(img_median,cv2.COLOR_BGR2RGB)



#Apply Filters

#Average
avg_blur=cv2.blur(img_avg_rgb,(5,5))
avg_blur_rgb=cv2.cvtColor(avg_blur,cv2.COLOR_BGR2RGB)

#Guassian
gaussian_blur=cv2.GaussianBlur(img_gaussian_rgb,(5,5),0)
gaussian_blur_rgb=cv2.cvtColor(gaussian_blur,cv2.COLOR_BGR2RGB)

#Median
median_blur=cv2.medianBlur(img_median_rgb,5)
median_blur_rgb=cv2.cvtColor(median_blur,cv2.COLOR_BGR2RGB)


#Display Results

plt.figure(figsize=(14,10))


plt.figure(figsize=(14, 10))

# ----- Average Blur -----
plt.subplot(3, 2, 1)
plt.imshow(img_avg_rgb)
plt.title("Original Image (Average Blur)")
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(avg_blur_rgb)
plt.title("After Average Blur")
plt.axis('off')


# ----- Gaussian Blur -----
plt.subplot(3, 2, 3)
plt.imshow(img_gaussian_rgb)
plt.title("Original Image (Gaussian Blur)")
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(gaussian_blur_rgb)
plt.title("After Gaussian Blur")
plt.axis('off')


# ----- Median Blur -----
plt.subplot(3, 2, 5)
plt.imshow(img_median_rgb)
plt.title("Original Image (Median Blur)")
plt.axis('off')

plt.subplot(3, 2, 6)
plt.imshow(median_blur_rgb)
plt.title("After Median Blur")
plt.axis('off')

plt.tight_layout()
plt.show()
```

Sharpening (Spatial Domain Filtering)

Sharpening is an image processing technique used to enhance edges and fine details in an image by emphasizing intensity differences between neighboring pixels.

Sharpening Using Kernel

Sharpening is performed using a kernel (mask).

The image is divided into small regions, and the kernel is applied to each region.

The kernel gives more importance (higher weights) to the center pixel and negative weights to surrounding pixels.

This highlights edges and details, making the image appear sharper.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np



img= cv2.imread("/content/blur-blurred-background-wallpaper.jpg",0)

# Check if image was loaded successfully
if img is None:
    print("Error: Could not load image. Please check the path and file integrity.")
else:
    kernel=np.array([[0,-1,0],
                     [-1,5,-1],
                     [0,-1,0]])

    sharpened=cv2.filter2D(img,-1,kernel)

    #Display Results

    plt.figure(figsize=(10,5))

    plt.subplot(1,3,1)
    plt.title("Original Image")
    plt.imshow(img,cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Sharpened Image")
    plt.imshow(sharpened,cmap='gray')
    plt.axis("off")
    plt.show()
```

Edge Detection

Edge detection is an image processing technique used to identify boundaries in an image where there is a sharp change in intensity.

1. Sobel Edge Detection

Sobel operator detects edges by computing the intensity gradient of the image.

Uses two kernels:

Horizontal (x)

Vertical (y)

Detects all intensity changes, including weak and strong edges.

Output may contain thicker edges and more noise.

2. Canny Edge Detection

Canny is a multi-stage, advanced edge detection algorithm.

Detects only strong (high-intensity) edges.

Produces thin and clean edges with less noise.

Steps in Canny

Noise reduction (Gaussian blur)

Gradient calculation

Non-maximum suppression

Double thresholding

Edge tracking by hysteresis

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np # Import numpy for potential use with gradients


img= cv2.imread("/content/HYF01MAYABAZAAR1.jpg",0)

#1. Sobel edge Detection

sobel_x=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sobel_y=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
sobel_xy=cv2.magnitude(sobel_x,sobel_y)


#2 Canny Edge Detection
canny_edges=cv2.Canny(img,100,200)

#Display Results

plt.figure(figsize=(12,6)) # Adjust figsize for better display of 4 images

plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(img,cmap='gray')
plt.axis("off")


plt.subplot(2,2,2)
plt.title("Sobel XY Edge")
plt.imshow(sobel_xy,cmap='gray') # Display Sobel XY result
plt.axis("off")

plt.subplot(2,2,3)
plt.title("Canny Edges")
plt.imshow(canny_edges,cmap='gray') # Display Canny Edges result
plt.axis("off")

plt.tight_layout() # Adjust subplot params for a tight layout
plt.show()
```
